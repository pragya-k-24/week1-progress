### MODIS Dataset Summary (India: 2021–2023)
#### About MODIS:
The Moderate Resolution Imaging Spectroradiometer (MODIS) is a key NASA sensor aboard the Terra (launched 1999) and Aqua (launched 2002) satellites. It captures Earth observation data at a spatial resolution of 1 km, suitable for global fire monitoring and environmental studies.

MODIS data used in this project is sourced from NASA’s FIRMS (Fire Information for Resource Management System) and focuses on thermal anomalies and active fire detection.

## Satellite Characteristics:
Terra satellite (EOS AM) captures morning overpasses.

Aqua satellite (EOS PM) captures afternoon overpasses.

MODIS provides 2–4 observations per day, especially in mid-latitudes like India.

## Fire Detection Mechanism:
*MODIS uses contextual algorithms to detect thermal anomalies.*

*It evaluates each pixel using mid-infrared channels (Bands 21/22 for  fire detection and 31 for surface temperature).*

*The pixel is marked as one of: missing, cloud, water, non-fire, fire, or unknown.*




### Note on Accuracy:
MODIS NRT (Near Real-Time) data may have slightly lower geolocation accuracy, particularly from Aqua satellite due to orbit estimation delays. Errors can occasionally reach several kilometers.



### Use Cases for MODIS Fire Data:
Real-time wildfire alerts

Agricultural burn detection

Forest fire management

Hotspot pattern analysis in ecological studies

### Experimental Analysis: Comparing FRP from Terra vs Aqua

```python
# Checking which satellite captures higher frp fires
plt.figure(figsize=(8, 6))
sns.countplot(x='satellite', data=df)
plt.title('Fire Radiative Power Check Aqua v/s Terra')
plt.xlabel('satellite')
plt.ylabel('Fire Radiative Power')
plt.show()
```

- The FRP plot shows a comparision between the data collected by Terra and Aqua in terms of Fire Radiative Power.
- It is evident that Aqua(EOS PM) satellite that captures afternoon overpasses has come accross higher FRP fires than Terra.
- FRP is a key indicator of fire intensity and is often used to estimate the rate at which fuel is being consumed by the fire.
- further analysis by acq_time and fire type can help pin point when and what cause result in such high radiation fires, thus recognising a pattern by time and location. 

```python
avg_frp = df.groupby(['acq_time', 'satellite'])['frp'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_frp, x='frp', y='acq_time', hue='satellite', marker='o')
plt.title('Avg FRP by Hour (Aqua vs Terra)')
plt.xlabel('Average Fire Radiative Power')
plt.ylabel('Hour of Day')
plt.grid(True)
plt.show()
```

- Between 500 to 750 UTC, Aqua shows more high-FRP events (100–300), indicating a possible link between time of day and fire intensity.
- Both satellites have one rare pike each exceeding 475 FRP, indicating anomaly or extreme fires.
- These patterns may be influenced by satellite overpass time and solar heating(acc. to time), making PM hours more fire-prone.
- Further analysis by adding fire type could help identify the type of fires responsible for these high readings.

```python
#
# Get average FRP grouped by acq_time, satellite, and fire_type
avg_frp = df.groupby(['acq_time', 'satellite', 'type'])['frp'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=avg_frp,
    x='frp',
    y='acq_time',
    hue='satellite',
    style='type',
    alpha=0.7,
    s=100,
    edgecolor='black',   
    linewidth=0.5  
)
plt.title('Avg FRP by Hour with Fire Types (Aqua vs Terra)')
plt.xlabel('Average Fire Radiative Power')
plt.ylabel('Hour of Day (UTC)')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


```

- Between 500 to 750 UTC, Aqua records several high-FRP events (100–300), many of which are marked as vegetation fires (type 0 and 3). 
- This could suggest a pattern of vegetation fires intensifying post-sunrise, possibly due to rising temperature or human activity.
- Fire types 2 and 3 appear less frequently, but when present they're of mid-to-high FRP values, with potentially more intense or clearly defined fire sources.
- While this trend is visible, further data (like location or weather) would further refine the cause of these peaks.
